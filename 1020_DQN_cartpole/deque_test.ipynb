```python
from collections import deque
```

```python
D = deque()
print(D)
```

```python
D = deque([1,2,3])
print(D)
D.append(10)
D.append(20)
print(D)
```

```python
D = deque([1,2,3], maxlen=4)
print(D)
D.append(10)
D.append(20)
D.append(30)
print(D)
```

```python
buffer = deque(maxlen=5)  # 최대 5개만 저장
experiences = [
    ("state1", 0, 1.0, "state2", False),
    ("state2", 1, 0.5, "state3", False),
    ("state3", 0, -0.1, "state4", False),
    ("state4", 1, 1.0, "state5", False),
    ("state5", 0, 0.0, "state6", True),
    ("state6", 1, 1.0, "state7", False),
    ("state6", 0, -0.5, "state8", False)
]

for i, exp in enumerate(experiences):
    buffer.append(exp)
buffer
```

```python
import random
buffer = deque(maxlen=1000)

# 경험 추가
for step in range(5):
    state = f"s{step}"
    action = step % 2
    reward = random.random()
    next_state = f"s{step+1}"
    done = (step == 4)
    
    buffer.append((state, action, reward, next_state, done))
    print(f"Step {step}: 버퍼 크기 = {len(buffer)}")

# 충분히 쌓였으면 샘플링
batch_size = 3
if len(buffer) >= batch_size:
    print(f"\n버퍼가 {batch_size}개 이상 → 샘플링 가능!")
    batch = random.sample(buffer, batch_size)
    
    # numpy array로 변환 (실제 학습에서 사용)
    import numpy as np
    states = np.array([exp[0] for exp in batch])
    actions = np.array([exp[1] for exp in batch])
    rewards = np.array([exp[2] for exp in batch])
    
    print(f"States: {states}")
    print(f"Actions: {actions}")
    print(f"Rewards: {rewards}")
```
